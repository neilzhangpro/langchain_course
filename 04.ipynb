```python
# 使用loader来加载markdown文本
# use loader to load markdown text
from langchain_community.document_loaders import TextLoader

loader = TextLoader("asset/loader.md")
loader.load()
```

```python
#使用loader来加载cvs文件

from langchain_community.document_loaders import CSVLoader

#loader = CSVLoader(file_path="asset/loader.csv")
loader = CSVLoader(file_path="asset/loader.csv",source_column="Location")
data = loader.load()
print(data)
```

```python
#某个目录下，有excel文件，我们需要把目录下所有的xlxs文件加载进来
# under a directory, there are excel files, we need to load all the xlxs files under the directory
from langchain_community.document_loaders import DirectoryLoader

#目录下的.html和.rst文件不会被这种loader加载
# .html and .rst files under the directory will not be loaded by this loader
#loader = DirectoryLoader("目录地址",glob="指定加载说明格式的文件")
loader = DirectoryLoader(path="./asset/example/",glob="*.xlsx")
docs = loader.load()
len(docs)
```

```python
#使用loader来加载html文件
# use loader to load html file
from langchain_community.document_loaders import BSHTMLLoader
loader = BSHTMLLoader("asset/loader.html")
data = loader.load()
data
```

```python
#使用loader来加载json文件
# use loader to load json file
from langchain_community.document_loaders import JSONLoader
loader = JSONLoader(
    file_path = "asset/simple_prompt.json",jq_schema=".template",text_content=True
)
data = loader.load()
print(data)
```

```python
#loader加载pdf文件
# loader load pdf file
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("asset/loader.pdf")
pages = loader.load_and_split()
pages[0]
```
