### 从环境变量中读取密钥
### Read the key from the environment variable
- 注意：尽量将你的OpenAI Key存储在类似.env文件中，而不是明文暴露在代码里，这是一种基本的安全措施
- Note: Try to store your OpenAI Key in something like an .env file, rather than exposing it explicitly in code, as a basic safety measure!
******

```python

import os
from dotenv import load_dotenv
# Load environment variables from openai.env file
load_dotenv("asset/openai.env")

# Read the OPENAI_API_KEY from the environment
api_key = os.getenv("OPENAI_API_KEY")
api_base = os.getenv("OPENAI_API_BASE")
os.environ["OPENAI_API_KEY"] = api_key
os.environ["OPENAI_API_BASE"] = api_base
os.environ["SERPAPI_API_KEY"] = os.getenv("SERPAPI_API_KEY")
os.environ["ELEVEN_API_KEY"] = os.getenv("ELEVEN_API_KEY")
os.environ["AZURE_COGS_KEY"] = os.getenv("AZURE_COGS_KEY")
os.environ["AZURE_COGS_ENDPOINT"] = os.getenv("AZURE_COGS_ENDPOINT")
os.environ["AZURE_COGS_REGION"] = os.getenv("AZURE_COGS_REGION")
```

### Tookit
- tookit是langchain已经封装好的一系列工具，一个工具包是一组工具来组合完成特定的任务
### Tookit
- tookit is a set of tools that langchain has encapsulated, a toolkit is a set of tools to combine to accomplish a specific task

- AzureCogsFormRecognizerTool：从文档里提取文本
- AzureCogsSpeech2TextTool：语音到文本
- AzureCogsText2SpeechTool：文本到语音

Azure Cognitive Services https://portal.azure.com/#allservices
- AzureCogsFormRecognizerTool: extract text from documents
- AzureCogsSpeech2TextTool: speech to text
- AzureCogsText2SpeechTool: text to speech

```python
#创建toolkit
from langchain.agents.agent_toolkits import AzureCognitiveServicesToolkit

toolkit = AzureCognitiveServicesToolkit()
```

```python
[tool.name for tool in toolkit.get_tools()]
```

```python
#agent使用
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, AgentType
llm = ChatOpenAI(temperature=0,model="gpt-4-1106-preview")
agent = initialize_agent(
    tools=toolkit.get_tools(),
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)
```

```python
audio_file = agent.invoke("Tell me a joke and read it out for me.")
```

```python
from langchain_experimental.agents.agent_toolkits import create_python_agent
from langchain_experimental.tools import PythonREPLTool
from langchain_experimental.utilities import PythonREPL
from langchain.llms.openai import OpenAI
from langchain.agents.agent_types import AgentType
from langchain_openai import ChatOpenAI
```

```python
agent_executor = create_python_agent(
    llm=ChatOpenAI(temperature=0, model="gpt-4-1106-preview"),
    tool=PythonREPLTool(),
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
    agent_executor_kwargs={"handle_parsing_errors": True},
)
```

```python
agent_executor.invoke("What is the 10th fibonacci number?")
```

```python
agent_executor.invoke(
    """Understand, write a single neuron neural network in PyTorch.
Take synthetic data for y=2x. Train for 1000 epochs and print every 100 epochs.
Return prediction for x = 5"""
)
```

### SQL Database
- 使用SQLDatabaseChain构建的agent，用来根据数据库回答一般行动问题
### SQL Database
- Agent built with SQLDatabaseChain to answer general action questions based on database

```python
from langchain.agents import create_sql_agent
from langchain.agents.agent_toolkits import SQLDatabaseToolkit
from langchain.sql_database import SQLDatabase
from langchain_openai import OpenAI

from langchain.agents import AgentExecutor
from langchain.agents.agent_types import AgentType
from langchain_openai import ChatOpenAI
```

```python
db = SQLDatabase.from_uri("sqlite:///asset/Chinook.db")
toolkit = SQLDatabaseToolkit(db=db, llm=OpenAI(temperature=0))
```

```python
agent_executor = create_sql_agent(
    llm=ChatOpenAI(temperature=0, model="claude-3-opus-20240229"),
    toolkit=toolkit,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
    handle_parsing_errors=True,
)
```

```python
agent_executor.invoke("Describe the playlisttrack table")
```

```python
try:
    agent_executor.invoke("Describe the Playlist table")
except Exception as e:
    print(e)
```
